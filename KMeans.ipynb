```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs 
```

```python
## X : m(examples)*n(features) matrix of training examples 
## centroids : array of centroids 
## returns : closest(m,) containing closest centroid index (in centroids) for training examples 

def find_closest_centroids(X,centroids):
    
    closest = np.zeros(X.shape[0],dtype=int) 
    
    for i in range(X.shape[0]):
        distance_from_centroids = [] 
        
        for j in range(centroids.shape[0]): ## Distance from each centroid ## 
            distance_from_centroids.append(np.linalg.norm(X[i] - centroids[j])) 
            
        closest[i] = np.argmin(distance_from_centroids) ##Minimum distance index assigned 
   
    return closest  
          
```

```python
## Synthetic Clusters 
X,y = make_blobs(n_samples=500, centers=3, random_state=6)

plt.scatter(X[:,0],X[:,1],c=y,cmap='Set1')  
plt.xlabel('x1') 
plt.ylabel('x2') 
plt.show() 
```

```python
## a look at what the function does ## 
initial_centroids = np.array([[1,2],[3,3],[0,0]]) 
closez = find_closest_centroids(X,initial_centroids)  

closez 
```

```python
## Compute mean of assigned points for each centroid 
def compute_centroids(X,closest,num_centroids):
    
    features = X.shape[1] 
    new_centroids = np.zeros((num_centroids,features)) 
     
    for i in range(num_centroids):
        
        assigned_points = X[closest == i] ## That point in X which has the ith centroid assigned (Pretty cool) 

        mean = np.mean(assigned_points,axis=0)  
        new_centroids[i] = mean  
        
    return new_centroids 


# Find closest centroids using initial_centroids
idx = find_closest_centroids(X, initial_centroids)

K = 3
centroids = compute_centroids(X, idx, K)

print("The centroids are:", centroids)
            
```

```python
## Where they at ##

plt.scatter(X[:,0],X[:,1],c=y,cmap='Set1')  

for x in centroids:
    plt.scatter(x[0],x[1],c=K)   
    
plt.show() 
```

```python
## ONCE MORE  
idx2 = find_closest_centroids(X,centroids) 
centroids2 = compute_centroids(X,idx2,K)

## Where they at ##

plt.scatter(X[:,0],X[:,1],c=y,cmap='Set1')  

for x in centroids2:
    plt.scatter(x[0],x[1],c=K)   
    
plt.show() 

```
